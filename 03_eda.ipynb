```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"D:\UAE-Telecom-Churn-Analytics\data\processed\cleaned_uae_telecom_data.csv")

df.head()
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
sns.countplot(x='Churn', data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
churn_percent = df['Churn'].value_counts(normalize=True) * 100

print(churn_percent)
```

```python
sns.countplot(x='Emirate', data=df)

plt.title("Customers by Emirate")

plt.xticks(rotation=45)

plt.show()
```

```python
pd.crosstab(df['Emirate'], df['Churn']).plot(kind='bar', stacked=True)

plt.title("Churn by Emirate")

plt.show()
```

```python
sns.histplot(df['MonthlyCharges'], bins=30)

plt.title("Monthly Charges Distribution")

plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

plt.title("Monthly Charges vs Churn")

plt.show()
```

```python
sns.boxplot(x='Churn', y='Tenure', data=df)

plt.title("Customer Tenure vs Churn")

plt.show()
```

```python
sns.countplot(x='5G_Subscription', hue='Churn', data=df)

plt.title("5G Subscription vs Churn")

plt.show()
```

```python
sns.countplot(x='Premium_Customer_Tier', hue='Churn', data=df)

plt.title("Premium Customer Churn")

plt.show()
```

```python
sns.histplot(df['Ramadan_Usage_Spike_Percentage'], bins=30)

plt.title("Ramadan Usage Spike Distribution")

plt.xlabel("Ramadan Usage Spike Percentage")

plt.show()
```

```python
sns.boxplot(x='Churn', y='Ramadan_Usage_Spike_Percentage', data=df)

plt.title("Ramadan Usage Spike vs Churn")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12,8))

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```
