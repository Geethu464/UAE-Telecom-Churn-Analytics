```python
import pandas as pd
import sqlite3

# Load cleaned dataset
df = pd.read_csv(r"D:\UAE-Telecom-Churn-Analytics\data\processed\cleaned_uae_telecom_data.csv")

# Create database
conn = sqlite3.connect(r"D:\UAE-Telecom-Churn-Analytics\sql\uae_telecom.db")

# Export dataframe to SQL table
df.to_sql('telecom_customers', conn, if_exists='replace', index=False)

print("Database created successfully")
```

```python
import sqlite3

conn = sqlite3.connect(
    r"D:\UAE-Telecom-Churn-Analytics\sql\uae_telecom.db"
)

print("Connected Successfully!")
```

```python
query = """
SELECT *
FROM telecom_customers
LIMIT 5;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT COUNT(*) AS total_customers
FROM telecom_customers;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT Churn,
       COUNT(*) AS customer_count
FROM telecom_customers
GROUP BY Churn;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT Churn,
       ROUND(
           COUNT(*) * 100.0 /
           (SELECT COUNT(*) FROM telecom_customers),
           2
       ) AS churn_percentage
FROM telecom_customers
GROUP BY Churn;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT Emirate,
       COUNT(*) AS total_customers
FROM telecom_customers
GROUP BY Emirate
ORDER BY total_customers DESC;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT CustomerID,
       MonthlyCharges,
       Premium_Customer_Tier
FROM telecom_customers
ORDER BY MonthlyCharges DESC
LIMIT 10;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT "5G_Subscription",
       COUNT(*) AS total_customers
FROM telecom_customers
GROUP BY "5G_Subscription";
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT Premium_Customer_Tier,
       COUNT(*) AS customer_count
FROM telecom_customers
GROUP BY Premium_Customer_Tier;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT ROUND(AVG(MonthlyCharges),2)
AS average_monthly_revenue
FROM telecom_customers;
"""

result = pd.read_sql(query, conn)

result
```

```python
query = """
SELECT Emirate,
       Churn,
       COUNT(*) AS customer_count
FROM telecom_customers
GROUP BY Emirate, Churn
ORDER BY Emirate;
"""

result = pd.read_sql(query, conn)

result
```
